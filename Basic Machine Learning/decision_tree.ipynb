```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('decision_tree.csv')
df
```

```python
age_mapping = {
    '<21':1,
    '21-35':2,
    '>35':3
}
```

```python
df['Age'] = df['Age'].map(age_mapping)
df
```

```python
income_mapping = {
    'Low':1,
    'Medium':2,
    'High':3
}
```

```python
df['Income']=df['Income'].map(income_mapping)
df
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Buys'] = le.fit_transform(df['Buys'])
```

```python
df
```

```python
df = pd.get_dummies(df[['ID', 'Age', 'Income', 'Gender', 'Martial Status', 'Buys']])
```

```python
df
```

```python
X = df[['Age', 'Income', 'Gender_Female', 'Gender_Male', 'Martial Status_Married', 'Martial Status_Single']]
X
```

```python
y = df['Buys']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X, y)
print('Model Accuracy : ',model.score(X, y)*100)
```

```python
X_test = [[1, 1, 1, 0, 1, 0]]
y_pred = model.predict(X_test)
print('Prediction : ', y_pred)
```

```python
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
dot_data = export_graphviz(model, filled=True, rounded=True, class_names=['Buys', 'Does not Buys'], 
                           feature_names=['Age', 'Income', 'Gender_Female', 'Gender_Male', 'Martial Status_Married', 'Martial Status_Single'])
graph = graph_from_dot_data(dot_data)
graph.write_png('tree.png')
```
